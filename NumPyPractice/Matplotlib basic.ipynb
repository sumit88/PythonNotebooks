```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(5)
print(x)
```

```python
y = x
print(y)
```

```python
plt.plot(x,y,'o') # o is here to tell them big spots
plt.show()
```

```python
plt.plot(x,y,'o-') # o- shows connected lines
plt.show()
```

```python
plt.plot(x,y,'o--') # o- shows dashed connected 
plt.show()
```

```python
plt.plot(x,y,'o--') 
plt.plot(x,-y,'o--')
plt.title('y==x and y =-x') # shows tiltle
plt.show()
```

```python
N = 10
x = np.linspace(0,10, N) # creates a linear range . 
      # Is different from arange as it calculates step size
print(x)
y =x
```

```python
N = 11
x = np.linspace(0,10, N) # creates a linear range . 
      # Is different from arange as it calculates step size
print(x)
y =x

plt.plot(x,y,'o--') 
plt.axis('off') # doesnt show any legends
plt.show()
```

```python
y = np.logspace(0.1 , 1 , N) # creates a log range
print(y)
plt.plot(x,y,'o--') 
 # doesnt show any legends
plt.show()
```

```python
y = np.logspace(1 , 100 , N) # creates a log range
print(y)
plt.plot(x,y,'o--') 
 # doesnt show any legends
plt.show()
```

```python
y = np.geomspace(0.1 , 1000 , N) # geometrix space cannot contain zero
print(y)
plt.plot(x,y,'o--') 
 # doesnt show any legends
plt.show()
```
